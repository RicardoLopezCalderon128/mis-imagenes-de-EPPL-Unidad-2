*  📝 Redacción del Problema
* Desarrollar un programa en Python que permita a un estudiante calcular propiedades geométricas básicas de figuras circulares.
* El programa debe mostrar un menú interactivo con cuatro opciones: calcular el área de un círculo, calcular su perímetro, calcular el volumen de una esfera y salir.
* Cada cálculo se realizará mediante funciones independientes que soliciten al usuario el valor del radio, apliquen la fórmula correspondiente y muestren el resultado en pantalla.
* El programa deberá mantenerse en ejecución hasta que el usuario decida salir.

* 🔍 Redacción Congruente con el Problema
* El programa está diseñado con un enfoque procedural, evitando el uso de programación orientada a objetos para mantener la simplicidad.
* Se utiliza un bucle while que muestra continuamente el menú, junto con estructuras condicionales if, elif y else para dirigir el flujo según la elección del usuario.
* No se incluye validación de datos para enfocarse en la lógica básica y el manejo de funciones. El radio se solicita dentro de cada función correspondiente, garantizando independencia y claridad en cada cálculo.
* La interfaz se adorna con emojis para hacer la experiencia más amigable e intuitiva.
📐 Fórmulas a Utilizar
1. Área del círculo: A = π · r²
2. Perímetro del círculo: P = 2 · π · r
3. Volumen de la esfera: V = (4/3) · π · r³
* (Donde π ≈ 3.14159... y r es el radio)

```python
import math
# elaborado por Ricardo Lopez Calderon
# ✅ Función para calcular el área de un círculo
def areacirculo():
    print("\n📐 Cálculo del Área de un Círculo")
    radio = float(input("👉 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"🔵 El área del círculo es: {area:.2f}\n")

# ✅ Función para calcular el perímetro de un círculo
def perimetrocirculo():
    print("\n📏 Cálculo del Perímetro de un Círculo")
    radio = float(input("👉 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"🔵 El perímetro del círculo es: {perimetro:.2f}\n")

# ✅ Función para calcular el volumen de una esfera
def volumenesfera():
    print("\n🌐 Cálculo del Volumen de una Esfera")
    radio = float(input("👉 Ingresa el radio de la esfera: "))
    volumen = (4/3) * math.pi * radio ** 3
    print(f"🔵 El volumen de la esfera es: {volumen:.2f}\n")

# ✅ Función menú principal
def menu():
    while True:
        print("✨ MENÚ GEOMÉTRICO ✨")
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")
        opcion = input("🎯 Selecciona una opción (1-4): ")

        if opcion == "1":
            areacirculo()
        elif opcion == "2":
            perimetrocirculo()
        elif opcion == "3":
            volumenesfera()
        elif opcion == "4":
            print("\n👋 ¡Gracias por usar el programa! Hasta luego.")
            break
        else:
            print("\n⚠️ Opción no válida. Por favor, elige entre 1 y 4.\n")

# ▶️ Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
